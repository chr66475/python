Importing the data.

```python
# Load package/module for data importing/wrangling.
## Here, I'm loading the 'pandas' package/module and naming it 'pd.'
import pandas as pd

# Loads the raw_data obtained from Kaggle. Names the raw data file that is imported 'raw_data.'
raw_data = pd.read_csv("C:/Users/conno/Documents/repos/python/test_analysis/data/raw_data/Conditions_Contributing_to_COVID-19_Deaths__by_State_and_Age__Provisional_2020-2023.csv")


# Allows me to view the first ten rows of the dataframe titled 'raw_data.'
raw_data.head(10)
```

```python
# Creates a new dataframe called 'clean_data' from the dataframe of 'raw_data'. Also filters columns so that it only shows rows where 'State' == 'United States' and 'Condition Group' == 'Respiratory diseases' and 'Age Group' != 'Not stated.'
clean_data = raw_data.loc[(raw_data['State'] == 'United States') & (raw_data['Condition Group'] == 'Respiratory diseases') & (raw_data['Age Group'] != 'Not stated')]

# Drops unncessary columns ("Month", "Year", "ICD10_codes", "Flag", and "Number of Mentions"). 'axis = 1' tells python that you want to drop columns, not rows. 'inplace = True' tells python you just want to 'paste' over the existing or aformented dataframe instead of creating a new one.
clean_data.drop(columns=['Month', 'Year', 'ICD10_codes', 'Flag', 'Number of Mentions'], axis=1, inplace=True)

# Allows me to view the first ten rows of the dataframe titled 'clean_data.'
clean_data.head(10)
```

```python
# Exports this cleaned dataframe to a csv file.
clean_data.to_csv("C:/Users/conno/Documents/repos/python/test_analysis/data/clean_data/clean_data.csv")
```
